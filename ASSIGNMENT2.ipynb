```python
df=pd.DataFrame(data)
def remove_outliers_iqr(df, columns):
  for column in columns:
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
  return df
```

```python
df_cleaned = remove_outliers_iqr(df, ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'BMI', 'DiabetesPedigreeFunction', 'Age'])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_cleaned['Pregnancies'] = scaler.fit_transform(df_cleaned[['Pregnancies']])
df_cleaned['Glucose'] = scaler.fit_transform(df_cleaned[['Glucose']])
df_cleaned['BloodPressure'] = scaler.fit_transform(df_cleaned[['BloodPressure']])
df_cleaned['SkinThickness'] = scaler.fit_transform(df_cleaned[['SkinThickness']])
df_cleaned['Insulin'] = scaler.fit_transform(df_cleaned[['Insulin']])
df_cleaned['BMI'] = scaler.fit_transform(df_cleaned[['BMI']])
df_cleaned['DiabetesPedigreeFunction'] = scaler.fit_transform(df_cleaned[['DiabetesPedigreeFunction']])
df_cleaned['Age'] = scaler.fit_transform(df_cleaned[['Age']])
print(df_cleaned)
```

```python
x=df_cleaned.drop('Outcome',axis=1)
y=df_cleaned['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.8,random_state=42)
```

# ***LOGISTIC REGRESSION***

```python
#Using Logistic Regression for
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(x_train,y_train)
```

```python
y_pred = log.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print('accuracy_score=',accuracy_score(y_test,y_pred))
print('precision_score=',precision_score(y_test,y_pred))
print('recall_score=',recall_score(y_test,y_pred))
print('f1_score=',f1_score(y_test,y_pred))
#
```
