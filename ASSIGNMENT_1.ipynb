```python
import numpy as np
import pandas as pd
#Reading the data to python
data1=pd.read_csv("churn-data-v2.csv")
#listing the columns
data1.columns.tolist()


```

```python
df = pd.DataFrame(data1)
# Calculating the mean of each columns
column_means = df.select_dtypes(include='number').mean()
print("Mean of each column:")
print(column_means)
```

```python
#Checking the null values present in the data set
#counting the number of null values in each column
null_counts=df.isnull().sum()
print(null_counts)



```

```python
#visualization of the dataset
#histogram
import matplotlib.pyplot as plt


# Plot histogram
plt.hist(data1['satisfaction'],  edgecolor='black')
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("counts")
plt.show()
```

```python
import seaborn as sns
#Filtering the data
df_new = df.select_dtypes(include="number")
sns.heatmap(df_new.corr(), annot=True)
plt.xticks(rotation=45)

```

```python
#scatter plot
plt.figure(figsize=(10,10))
plt.scatter(df_new["average_montly_hours"],df_new["satisfaction"],c="k",marker="^")
plt.title("Average monthly hours Vs Satisfaction")
plt.xlabel("average_montly_hours")
plt.ylabel("satisfaction")
plt.show()
```
