```python
#CHECKING FOR OUTLIERS
plt.boxplot(df['ApplicantIncome'])
```

```python
#defining outlier removal function function
def remove_outliers_iqr(df, columns):
  for column in columns:
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
  return df
```

```python
df_cleaned = remove_outliers_iqr(df, ['ApplicantIncome','CoapplicantIncome','LoanAmount'])
```

#MINMAX SCALING



```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
df_cleaned['ApplicantIncome']=ms.fit_transform(df_cleaned[['ApplicantIncome']])
df_cleaned['CoapplicantIncome']=ms.fit_transform(df_cleaned[['CoapplicantIncome']])
df_cleaned['LoanAmount']=ms.fit_transform(df_cleaned[['LoanAmount']])
df_cleaned['Loan_Amount_Term']=ms.fit_transform(df_cleaned[['Loan_Amount_Term']])
```

#SPLITTING THE DATA

```python
x=df_cleaned.drop('Loan_Status',axis=1)
y=df_cleaned['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.8,random_state=42)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score
```

#RANDOM FOREST


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
rf=RandomForestClassifier()
parameters = {
    'n_estimators': [50, 100, 200],
    'criterion' :['gini','entropy'],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'n_jobs' : [2]
}
rf_cv = GridSearchCV(rf,param_grid=parameters,scoring='accuracy',n_jobs=2,cv=5)
rf_cv.fit(x_train,y_train)
```

```python
rf_cv.best_params_
```

```python
rf=rf_cv.best_estimator_
print(rf_cv.best_score_)
```

```python
y_pred_rf=rf.predict(x_test)
print('accuracy score=',accuracy_score(y_test,y_pred_rf))
print('precision_score=',precision_score(y_test,y_pred_rf))
print('recall_score=',recall_score(y_test,y_pred_rf))
print('f1_score=',f1_score(y_test,y_pred_rf))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred_dt=dt.predict(x_test)
```
